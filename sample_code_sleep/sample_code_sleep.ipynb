### Reading and visualising the mini-apple dataset (derived from Apple Watch Sleep Study[1])
##### Olivia Walch, Yitong Huang, Daniel Forger, and Cathy Goldstein, "Sleep stage prediction with raw acceleration and photoplethysmography heart rate data derived from a consumer wearable device", Sleep, 2019


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# for subject with a certain id 
#id = '844359'
id = '46343'

accelerometer_file = id + '_cleaned_resampled_1_hz.out'
heart_rate_file = id + '_cleaned_hr.out'
sleep_stages_file = id + '_labeled_sleep.txt'

print(accelerometer_file)
print(heart_rate_file)
print(sleep_stages_file)
```

```python
with open(accelerometer_file, 'r') as file:
    print('Now reading ' +accelerometer_file+ ' ...')
    # Read all lines of the file
    xyz_list = file.readlines()
    
xyz_mat = []
for row in xyz_list[1:]:
    # Strip newline characters, split by whitespace, and convert to floats
    xyz_mat.append([float(value) for value in row.strip().split()])

#print(xyz_mat)
```

```python
# Real accelerometer data (time, X, Y, Z)
data = xyz_mat

# Extract time, X, Y, Z from the data
time = [row[0] for row in data]
x = [row[1] for row in data]
y = [row[2] for row in data]
z = [row[3] for row in data]

# Plot the data
plt.figure(figsize=(12, 4))
plt.plot(time, x, label='accelerometer-x (1Hz)')
plt.plot(time, y, label='accelerometer-y (1Hz)')
plt.plot(time, z, label='accelerometer-z (1Hz)')

# Add labels, title, and legend
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s)')
plt.title('Movement for One-Night Sleep')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```

```python
with open(heart_rate_file, 'r') as file:
    # Read all lines of the file
    hr_list = file.readlines()
    
hr_mat = []
for row in hr_list:
    # Strip newline characters, split by whitespace, and convert to floats
    hr_mat.append([float(value) for value in row.strip().split()])
    
#print(hr_mat)
```

```python
# Real heart data (time, heart_rate)
data = hr_mat

# Extract time, heart_rate
time = [row[0] for row in data]
hr = [row[1] for row in data]


# Plot the data
plt.figure(figsize=(12, 3))
plt.plot(time, hr, label='Heart rate (Beats per Minute)')

# Add labels, title, and legend
plt.xlabel('Time (s)')
plt.ylabel('Heart Rate')
plt.title('Heart Rate for One-Night Sleep')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
with open(sleep_stages_file, 'r') as file:
    # Read all lines of the file
    sleep_list = file.readlines()
sleep_mat = []
for row in sleep_list:
    # Strip newline characters, split by whitespace, and convert to floats
    sleep_mat.append([float(value) for value in row.strip().split()])
```

```python
data = sleep_mat

time = [row[0] for row in data]
stage = [row[1] for row in data]


# Plot the data
plt.figure(figsize=(12, 3))
plt.plot(time, stage, label='-1:Invalid label; 0:Wake, 1:N1, 2:N2, 3:N3, REM:5')

# Add labels, title, and legend
plt.xlabel('Time (s)')
plt.ylabel('Sleep Stage')
plt.title('Five Stages for One-Night Sleep')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
#plt.savefig('stage.png', bbox_inches='tight')
```
